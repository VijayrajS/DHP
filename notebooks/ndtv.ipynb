```python
driver.get('https://www.ndtv.com/topic/caa')

```

```python
but = driver.find_element_by_class_name('loadmore_btn')
article_links = []
while but:
    c = driver.find_element_by_id('news_list')
    x = c.find_elements_by_tag_name('div')
    pages = x[2].find_elements_by_tag_name('li')
    for i in pages:
        link = i.find_element_by_tag_name('a').get_attribute('href')
        if link not in article_links:
            article_links.append(link)
    but = driver.find_element_by_class_name('loadmore_btn')
    but.click()
    print(len(article_links))
```
