```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
options = Options()
options.binary_location = "/bin/chromium"
driver = webdriver.Chrome(chrome_options=options, executable_path='/run/media/badwolf/Windows/Users/Urja Nadibail/acads/chromedriver')
```

```python
driver.get('https://www.thequint.com/search?q=CAA')

```

```python
c = driver.find_elements_by_class_name('card-elements')
x = driver.find_elements_by_tag_name('button')
for i in x:
    if i.text == "LOAD MORE":
        butt = i
articles = []
import time
while(butt):
    c = driver.find_elements_by_class_name('card-elements')
    print("Length", len(c))
    for i in c:
        if i.find_element_by_tag_name('a').get_attribute('href') not in articles:
            articles.append(i.find_element_by_tag_name('a').get_attribute('href'))
    element_to_hover_over = butt

    hover = ActionChains(driver).move_to_element(element_to_hover_over)
    hover.perform()
    try:
        butt.click()
    except:
        time.sleep(2)
    x = driver.find_elements_by_tag_name('button')
    for i in x:
        if i.text == "LOAD MORE":
            butt = i
    print(len(articles))
```

```python
import requests
from bs4 import BeautifulSoup
url = articles[1]
r1 = requests.get(url)
coverpage = r1.content
soup1 = BeautifulSoup(coverpage, 'html5lib')
```

```python
x = soup1.find('div', class_='story-article__body container').text
soup1.find('time')['datetime']
fin_arts = []
for i in articles:
    url = i
    r1 = requests.get(url)
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    d = {}
    d['link'] = i
    d['text'] = soup1.find('div', class_='story-article__body container').text
    d['time'] =    soup1.find('time')['datetime']
    fin_arts.append(d)
```

```python
import pickle 
with open('the_quint_caa.pickle', 'wb') as handle:
    pickle.dump(fin_arts, handle)
```
